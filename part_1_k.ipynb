```python
#for html
import requests
from bs4 import BeautifulSoup
# Import date class and datetime class from datetime module
from datetime import date 
from  datetime import datetime as datetime_func
#for DataFrame
import pandas as pd  
```

```python

def get_flat_info(url):  
    response = requests.get(url)
    if response.status_code == 200:
        print("Success")
    else:
        print("Failure")
        return None
    try:
        html = requests.get(url)
        soup = BeautifulSoup(html.content,'html.parser')

        # town and neigh. 
        #print(soup.find('div',class_="card-body p-3").find('p').get_text())

        #rooms number, m^2 and floor
        #print(soup.find('div', class_="d-flex flex-wrap").findAll('span'))
        #print(soup.find('div', class_="d-flex flex-wrap").findAll('span')[0].get_text().strip())
        #print(soup.find('div', class_="d-flex flex-wrap").findAll('span')[1].get_text().strip())
        #print(soup.find('div', class_="d-flex flex-wrap").findAll('span')[2].get_text().strip())

        #price and address
        price=float(0)
        pr_data=soup.find('div', class_="d-flex justify-content-between").findAll('h2')
        #print("address",pr_data[0].get_text().strip())
        pr_price=pr_data[1].get_text().split()[0].strip()
        len_price=len(pr_price)
        pr_price=list(pr_price)
        for i in range(len_price):
            if (pr_price[i]==','):
                #pr_price[i]='.' 
                pr_price[i]=''
        pr_price=''.join(pr_price)       
        #price=float(pr_price)*1000
        price=float(pr_price)
        #print("price:",price)

        #appartment"s properties
        property_type='None'
        neighborhood='None'
        address='None'

        room_num=float(0) 
        floor=int(-5)
        area=int(0)
        garden_area=int(0)
        days_to_enter=int(-1)
        num_of_payments=int(0)
        monthly_arnona=int(0)
        building_tax=int(0)
        total_floors=int(0)

        #fill in appartment's properties
        params=soup.find('table', class_ ="table table-sm mb-4").findAll('tr')
        #print(type(params))
        for i in range(len(params)):
            parm_tx=params[i].get_text().split()[0]
            #print(parm_tx)
            if (parm_tx == "פרטי"):
                #print(params[i].get_text().split()[2:],type(params[i].get_text().split()[2:]))
                property_type=' '.join(params[i].get_text().split()[2:])
                #print("property_type",property_type)
            elif parm_tx == "שכונה":
                #print(params[i].get_text().split()[1:],type(params[i].get_text().split()[1:]))
                neighborhood=' '.join(params[i].get_text().split()[1:])
                #print("neighborhood",neighborhood)
            elif parm_tx == "כתובת":
                #print(params[i].get_text().split()[1:],type(params[i].get_text().split()[1:]))
                address=' '.join(params[i].get_text().split()[1:])
                #print("address",address)    
            elif parm_tx == "חדרים":
                #print(params[i].get_text().split()[1],type(params[i].get_text().split()[1]))
                room_num=float(params[i].get_text().split()[1])
                #print("room_num",room_num)
            elif parm_tx=="קומה":
                #print(params[i].get_text().split()[1],type(params[i].get_text().split()[1]))
                #print(params[i].get_text().split()[3],type(params[i].get_text().split()[3]))
                if (params[i].get_text().split()[1]=="קרקע" and params[i].get_text().split()[3]=="קרקע"):
                    floor=int(0)
                    total_floors=int(1)
                    #print("floor ",floor," total_floors ",total_floors)
                elif (params[i].get_text().split()[1]!="קרקע" and params[i].get_text().split()[3]!="קרקע"):  
                    floor=int(params[i].get_text().split()[1])
                    total_floors=int(params[i].get_text().split()[3])
                    #print("floor ",floor," total_floors ",total_floors)
                elif (params[i].get_text().split()[1]=="קרקע" and params[i].get_text().split()[3]!="קרקע"):
                    floor=int(0)
                    total_floors=int(params[i].get_text().split()[3])
                    #print("floor ",floor," total_floors ",total_floors)
                elif (params[i].get_text().split()[1]!="קרקע" and params[i].get_text().split()[3]=="קרקע"):
                    floor=int(params[i].get_text().split()[1])  
                    total_floors=int(1)
                    #print("floor ",floor," total_floors ",total_floors)           
            elif parm_tx=="שטח":
                #print(params[i].get_text().split()[2],type(params[i].get_text().split()[2]))
                if (params[i].get_text().split()[1]=="בנוי"):
                    area=int(params[i].get_text().split()[2]) 
                    #print("area ",area)
                elif (params[i].get_text().split()[1]=="גינה"): 
                    garden_area=int(params[i].get_text().split()[2])
                    #print("garden area ",garden_area)
            elif parm_tx=="תאריך": 
                #print(params[i].get_text().split()[2],type(params[i].get_text().split()[2])) 
                if (params[i].get_text().split()[2]=="מיידית"):
                    days_to_enter=int(0)
                    #print("days_to_enter",days_to_enter)
                else: 
                    date_d=params[i].get_text().split()[2]
                    d = datetime_func.strptime(date_d, "%d/%m/%Y")
                    #date_for_ent
                    pr_d2 = datetime_func.strftime(d, '%Y-%m-%d')
                    d2=datetime_func.strptime(pr_d2, '%Y-%m-%d')
                    #print(d2,type(d2)) 
                    #current time
                    # importing datetime module for now()
                    import datetime
                    d1 = datetime.datetime.now()
                    #print(d1,type(d1))
                    #we want to know if all the parameters are legal
                    if (d1.year>d2.year):
                        days_to_enter=int(0)
                        #print("Delta ",days_to_enter)
                    elif (d1.year==d2.year):  
                        if (d1.month<d2.month):
                            delta=d2-d1
                            days_to_enter=delta.days
                            #print("Delta ",days_to_enter)
                    elif (d1.month==d2.month):
                        if (d1.day<=d2.day):
                            delta=d2-d1
                            days_to_enter=delta.days
                            #print("Delta",days_to_enter)
                        else:
                            days_to_enter=int(0)
                            #print("Delta",days_to_enter)
                    elif (d1.month>d2.month): 
                        days_to_enter=int(0)
                        #print("Delta",days_to_enter)
                    elif (d1.year<d2.year):  
                        delta=d2-d1
                        days_to_enter=delta.days
                        #print("Delta",days_to_enter)
            elif parm_tx=="תשלומים":
                #print(params[i].get_text().split()[2],type(params[i].get_text().split()[2]))    
                num_of_payments=int(params[i].get_text().split()[2])
                #print("num_of_payments",num_of_payments)        
            elif parm_tx=="ארנונה": 
                #print(params[i].get_text().split()[2],type(params[i].get_text().split()[2])) 
                monthly_arnona=int(params[i].get_text().split()[2])
                #print("monthly_arnona",monthly_arnona)
            elif parm_tx=="ועד": 
                #print(params[i].get_text().split()[3],type(params[i].get_text().split()[3])) 
                building_tax=int(params[i].get_text().split()[3])
                #print("building_tax",building_tax)


        # class with all apartment's elements, we will put to the params_in
        params_in=soup.find('div', class_="card-icons flex-wrap d-flex h-100").findAll('div')

        #loop for every element in appartment, 0-none, 1-Yes.
        flag=[]
        for i in range(len(params_in)):
            if (len(params_in[i]['class'])==3):
                flag.insert(i,1)
            elif (len(params_in[i]['class'])==4):
                flag.insert(i,0)
            # elif:
            #     print("problem")
            # break
        # list with status of all elements in the appartment
        #print(flag)

        # 1 yes, 0 no furniture
        is_furnished=flag[1]
        #print(params_in[1].get_text().strip(),"is_furnished",is_furnished)

        # 1 yes, 0 no air conditioner
        ac=flag[2]
        #print(params_in[2].get_text().strip(),"ac",ac)

        # 1 yes, 0 no parking
        has_parking=flag[3]
        #print(params_in[3].get_text().strip(),"has_parking",has_parking)

        # 1 yes, 0 no safe room
        has_safe_room=flag[4]
        #print(params_in[4].get_text().strip(),"has_safe_room",has_safe_room)

        # 1 yes, 0 no balcon
        has_balcon=flag[5]
        #print(params_in[5].get_text().strip(),"has_balcon",has_balcon)

        # 1 yes, 0 no handicap
        handicap=flag[6]
        #print(params_in[6].get_text().strip(),"handicap",handicap)

        # 1 yes, 0 no bars
        has_bars=flag[7]
        #print(params_in[7].get_text().strip(),"has_bars",has_bars)

        # 1 yes, 0 no elevator
        elevator=flag[8]
        #print(params_in[8].get_text().strip(),"elevator",elevator)

        # 1 yes, 0 no storage
        has_stotsge=flag[9]
        #print(params_in[9].get_text().strip(),"has_stotsge",has_stotsge)

        # 1 yes, 0 no renovated
        is_renovated=flag[11]
        #print(params_in[11].get_text().strip(),"is_renovated",is_renovated)

        num_of_images=int(0)
        description=str('None')

        #number of images
        #print(len(list_image_1),type(list_image_1))
        pr_list_images=soup.find('div',class_="col-12 d-flex mt-3 justify-content-center flex-wrap")
        if (pr_list_images):
            list_images=pr_list_images.findAll('div')
            #print(len(list_images),type(list_images))
            num_of_images=len(list_images)
            #print(num_of_images,type(num_of_images))
        else:
            num_of_images=int(0)
            #print("Number of images", num_of_images)

        #description of appartment
        pr_description=soup.findAll('p', class_="text-word-break")
        if (pr_description):
            for i in range(len(pr_description)):
                if (description=='None'):
                    description_1=pr_description[i].get_text().strip()
                    #print(description_1)
                else:    
                    description_1+=pr_description[i].get_text().strip()
                    #print(description_1)
            #print(description_1)
        else:
            description_1=description
            #print(description_1)
        description = " ".join(description_1.split())
        #print(description, type(description), len(description))
        # dictionary with variables like columes names   

        flat_dict = dict(property_type=property_type, neighborhood=neighborhood, address=address, room_num=room_num, floor=floor, area=area, garden_area=garden_area, days_to_enter=days_to_enter, num_of_payments=num_of_payments, monthly_arnona=monthly_arnona, building_tax=building_tax, total_floors=total_floors, description=description, has_parking=has_parking, has_stotsge=has_stotsge, elevator=elevator, ac=ac, handicap=handicap, has_bars=has_bars, has_safe_room=has_safe_room, has_balcon=has_balcon, is_furnished=is_furnished, is_renovated=is_renovated, price=price, num_of_images=num_of_images )
        return(flat_dict)
    except:
        return None
    
```

```python

def df_of_flats():
    url = input('Enter all neighborhoods in Tel-Aviv that you want to check- ')
    response = requests.get(url)
    if response.status_code == 200:
        print("Success")
    else:
        print("Failure")
        return None
    try:
        html = requests.get(url)
        soup = BeautifulSoup(html.content,'html.parser')

        #number of pages
        num_pg=int(soup.find('div', class_="d-flex align-items-baseline").find('h6').get_text().replace('(','').replace(')','').split()[3])
        #print(num_pg,type(num_pg))


        #for every page, take data from every announcement on the page
        link_start="https://www.ad.co.il/ad/"
        link_next="&pageindex=" #for next page
        k=0 #for every page
        list_of_flat=[] #list with dictionaries of flats
        for i in range(num_pg):
            if (i==0): #first page
                announ_list=soup.find('div', class_="cards-wrap s m l", id="cards").findAll('div', class_="card-body p-md-3") #"card-block hovered")
                #print(type(announ_list),len(announ_list))
                len_ann_pg=len(announ_list)
                data=soup.findAll('div', class_="card-block")
                #print(data)
                #for every announcement on the page
                for j in range(len_ann_pg):
                    id=data[j].get('data-id')
                    #print(id)
                    link= link_start+id
                    #print(link)
                    list_of_flat.insert((k+j),get_flat_info(link))
            else: #all pages except first
                url_1=url+link_next+ind
                #print(url_1)
                html_1 = requests.get(url_1)
                soup_1 = BeautifulSoup(html_1.content,'html.parser')
                announ_list=soup_1.find('div', class_="cards-wrap s m l", id="cards").findAll('div', class_="card-body p-md-3") #"card-block hovered")
                #print(type(announ_list),len(announ_list))
                len_ann_pg=len(announ_list)
                data=soup_1.findAll('div', class_="card-block")
                #print(data)
                #for every announcement on the page
                for j in range(len_ann_pg):
                    id=data[j].get('data-id')
                    #print(id)
                    link= link_start+id
                    #print(link)
                    list_of_flat.insert((k+j),get_flat_info(link))    
            k=k+len_ann_pg #list index
            ind=str(i+2) #page index
            #print("##################################_page_num=",i+1)
        #list of dictionaries to DataFrame
        df = pd.DataFrame.from_records(list_of_flat)
        #print(df)
        #csv file
        df.to_csv('flat_out.csv', index=False) 
    except:
        return None
```
